# GUI Programming
### Graphical user interface 13.1
**Concept**: A graphical user interface allows the user to interact with the operating\
system and other programs using graphical elements such as icons, buttons, and dialogue boxes.

One part of the user interface consist of hardware devices, such as the keyboard and\
the video display. Another part lies in the way that the computer's operating system\
accepts commands from the user. For years, the only way that the user could interact\
with the operating system was through a command line interface. A command line interface\
typically displays a prompt, and the user types a command, which is then executed.

Many users find command line interfaces difficult to use because there are many commands\
to be learned, and each command has its own syntax, much like a programming statement.

A new type of interface known as a graphical user interface came into use in commercial\
operating systems. A graphical user interface allows user to interact with the operating\
system and other programs using graphical elements on the screen. GUI popularized the use\
of the mouse as an input device, instead the user having to type commands, GUI allows the\
user to point at graphical elements and use the mouse to click them.

Much of the interaction with GUI is done through dialogue boxes, which are small windows\
that display information and allows the user to perform actions.

### GUI programs are Event driven
A command line interface, determines the order in which things happen. The user has no\
choice but to enter data in the order that is requested. In a GUI environment, the user\
determines the order in which things happen, because GUI programs must respond to the\
actions of the user, is said that they're event-driven.

The simplest GUI program that we can demonstrate is one that one displays an empy window,\

````python
# This program displays an empty window

import tkinter

def main():
    """
    Creates a window.
    
    :return: None
    """
    # Create the main window widget 
    main_window = tkinter.Tk()
    
    # Enter the tkinter main loop
    tkinter.mainloop()

# Call the main function  
main()
````

Line 15 call the mainloop function which runs like an infinite loop until you close the main window. Most programmers prefer to take an object-oriented approach when writing GUI program. It is common to write a class with an __init__ method that builds the GUI. When an instance of a class is created the GUI is displayed on screen. The program below is an example of this:


````python
# This program displays an empty window
import tkinter

class MyGUI:
    """
    Description
    ===========
    Creates an empty window.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget 
        main_window = tkinter.Tk()
            
        # Enter the tkinter main loop
        tkinter.mainloop()

# Creates an instance of the MyGUI class
my_gui = MyGUI
````

### Display Text with Label Widgets 13.3
**Concept**: You can use the Label widget to display text in a window.

You can use the Label to display a single line of text in a window. To make a Label\
you need to make an instance of the tkinter Module Label class.


````python
# This program displays a label with text

import tkinter

class GUI:
    """
    Description
    ===========
    Creates a window, with a single line of text inside.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk()
        
        # Create a label widget containing the 
        # text 'Hello World'
        self.__label = tkinter.Label(self.__main_window,
                                     text= 'Hello World')
        
        # Calling the label widget's pack method
        self.__label.pack()
        
        # Enter the tkinter main loop
        tkinter.mainloop()

# Creates an instance of the MyGUI class
my_gui = MyGUI
````

This statement creates a label widget and assigns it to self.__label. The first argument,\
self.__main_window, is a reference to the root widget. THis simply specifies that we want\
the Label widget to belong to the root widget, the second argument, text= 'Hello World',\
specifies the text that we want to be displayed in the label. Line 25 call the Label widgets\
pack method. The pack method determine where a widget should be positioned and makes the \
widget visible when the main window is displayed.


````python
# This program displays two labels with text 

import tkinter

class GUI:
    """
    Description
    ===========
    Creates a window, with two lines of text inside.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk
        
        # Creates two label widgets
        self.label1 = tkinter.Label(self.__main_window,
                                    text= 'Hello World')
        self.label2 = tkinter.Label(self.__main_window,
                                    text= 'This is my GUI program')
        
        # Call both the label widget's pack method
        self.label1.pack()
        self.label2.pack()
        
         # Enter the tkinter main loop
        tkinter.mainloop()
        
# Creates an instance of the MyGUI class
my_gui = MyGUI
````

Notice the two label widgets are stacked with one on top of the other. We can change\
this layout by specifying an argument to pack method.


````python
# This program displays two labels with text 

import tkinter

class GUI:
    """
    Description
    ===========
    Creates a window, with two lines of text on the left side of it.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk
        
        # Creates two label widgets
        self.label1 = tkinter.Label(self.__main_window,
                                    text= 'Hello World')
        self.label2 = tkinter.Label(self.__main_window,
                                    text= 'This is my GUI program')
        
        # Call both the label widget's pack method
        self.label1.pack(side= 'left')
        self.label2.pack(side= 'left')
        
         # Enter the tkinter main loop
        tkinter.mainloop()
        
# Creates an instance of the MyGUI class
my_gui = MyGUI
````

We call the label widget's pack method passing side= 'left' as an argument. This specifies\
that the widget should be positioned as far to the left as possible in the parent widget.\
The valid side argument you can pass  to the pack method are side= 'top', side= 'bottom',\
side='left', side='right'.

### Organizing Widgets with Frames 13.4
**Concept**: A Frame is a container that can hold other widgets. You can use frames\
to organize the widgets in a window.

frames are useful organizing and arranging groups of widgets in a window. for example you can place a set of widgets in one frame and another set in another frame.


````python
import tkinter

class GUI:
    """
    Description
    ===========
    Creates a window, with two lines of text on buttom and top of it.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk
        
        # Create two frames, one for the top of the
        # window, and one for the bottom
        self.top_frame = tkinter.Frame(self.__main_window)
        self.bottom_frame = tkinter.Frame(self.__main_window)
        
        # Create three label widgets for the 
        # top frame
        self.label1 = tkinter.Label(self.top_frame, 
                                    text= 'Winken')
        self.label2 = tkinter.Label(self.top_frame, 
                                    text= 'Blinken')
        self.label3 = tkinter.Label(self.top_frame, 
                                    text= 'Nod')
        # Pack the labels that are in the top frame
        # Use the side= 'top' argument to stack
        self.label1.pack(side= 'top')
        self.label2.pack(side= 'top')
        self.label3.pack(side= 'top')
        
        # Create three label widgets for the 
        # bottom frame
        self.label4 = tkinter.Label(self.bottom_frame, 
                                    text= 'Winken')
        self.label5 = tkinter.Label(self.bottom_frame, 
                                    text= 'Blinken')
        self.label6 = tkinter.Label(self.bottom_frame, 
                                    text= 'Nod')
        # Pack the labels that are in the bottom frame
        # Use the side= 'left' argument to stack
        self.label4.pack(side= 'left')
        self.label5.pack(side= 'left')
        self.label6.pack(side= 'left')
        
        # Yes, we have to pack the frames
        self.top_frame.pack()
        self.bottom_frame.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
# Creates an instance of the MyGUI class
my_gui = MyGUI
````

### Button Widgets and Info Dialogue Boxes 13.5
**Concept**: You can use the Button widget to create a standard button in a window. When the user clicks the butttom, a specific function or method is called. And info dialogue box is a simple window that displays a message  to the user, and has an OK button that dismisses the dialogue box. You can use the tkinter.messagebox module's showinfo function to display an info dialogue box.

When you create a button widget you can specify the text that is to appear on the face of the face of the button and name of the callback function. A callback function is a function or method that executes when the user clicks the button. 

Here's the general format of a showinfo function:
```
tkinter.messagebox.showinfo(title, messsage)
```


````python
import tkinter
import tkinter.messagebox

class GUI:
    """
    Description
    ===========
    Creates a window, with a button giving the user an option to click it.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk
        
        # Create the button widget. The text 'Click me'
        # should appear on the face of the Button. The
        # do_something method should be executed when
        # the user clicks the button
        self.my_button = tkinter.Button(self.__main_window,
                                        text= 'Click me',
                                        command= self.do_something)
        # Pack the button
        self.my_button.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
    # The do_something method is the callback function
    # for the button widget
    
    def do_something(self):
        tkinter.messagebox.showinfo('Response', 
                                    'Thanks for clicking the button')
        
# Creates an instance of the MyGUI class
my_gui = MyGUI
````

### Creating a Quit Button
GUI programs usually have a quit button that closes the program when the user clicks it. To create a quit button you simply create a button widget that calls the root widget's destroy method as a callback function.


````python
import tkinter
import tkinter.messagebox

class GUI:
    """
    Description
    ===========
    Creates a window, with two buttons giving the user an option to click one or quit th e program.
    
    Usage
    =====
    >>> my_gui = MyGUI
    """
    def __init__(self):
        # Create the main window widget
        self.__main_window = tkinter.Tk
        
        # Create the button widget. The text 'Click me'
        # should appear on the face of the Button. The
        # do_something method should be executed when
        # the user clicks the button
        self.my_button = tkinter.Button(self.__main_window,
                                        text= 'Click me',
                                        command= self.do_something)
        
        # Create a Quit button. When this button is clicked 
        # the root widget's destroy method is called. 
        # (The Main window variable references the root widget
        # so the callback function is self.__main_window.destroy())
        self.quit_button = tkinter.Button(self.__main_window,
                                        text= 'Quit',
                                        command= self.__main_window.destroy())
        
        # Pack the button
        self.my_button.pack()
        self.quit_button.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
    # The do_something method is the callback function
    # for the button widget
    
    def do_something(self):
        tkinter.messagebox.showinfo('Response', 
                                    'Thanks for clicking the button')
        
# Creates an instance of the MyGUI class
my_gui = MyGUI
````

### Getting Input with the Entry Widget 13.6
**Concept**: An Entry is a rectangular area that a user can type input into. You use the Entry widgets get method to retrieve data that has been typed into the widget.

Entry widgets are used to gather input in a GUI program. A program will have one or more Entry widgets in window alon with a button that the user clicks to submit that they've typed into the Entry widget.

You use the Entry widgets get method to retrieve data that has been typed into the widget. The get method returns a string, so it'll have to be converted to the correct data type if the Entry widget is used for numeric input.

To demonstrate, we will look at a program that allows the user to enter a distance in kilometers into an Entry widget then click a button to see the distance converted to miles.


````python
# This program converts distance in kilometers
# to miles. The result is displayed in an info
# dialogue box.

import tkinter
import tkinter.messagebox

class KiloConverterGUI:
    def __init__(self):
        # Create the main window widget
        self.main_window = tkinter.TK()
        
        # Create two frames to group widgets
        self.top_frame = tkinter.Frame(self.main_window)
        self.bottom_frame = tkinter.Frame(self.main_window)
        
        # Create the widgets for the top frame
        self.prompt_label = tkinter.Label(self.top_frame, 
                                          text= 'Enter a distance in kilometers')
        self.kilo_Entry = tkinter.Entry(self.top_frame, 
                                        width= 10)
        
        # Pack the top frame's widgets
        self.prompt_label.pack(side= 'left')
        self.kilo_Entry.pack(side= 'left')
        
        # Create the button widgets for the bottom frame
        self.calc_button = tkinter.Button(self.bottom_frame,
                                          text= 'Convert',
                                          command= self.convert)
        
        self.quit_button =  tkinter.Button(self.bottom_frame,
                                          text= 'Quit',
                                          command= self.main_window.destroy)
         # Pack the buttons
        self.calc_button.pack(side= 'left')
        self.quit_button.pack(side= 'left')
        
        # Pack the Frames
        self.top_frame.pack()
        self.bottom_frame.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
    # The convert method is a callback function for
    # the calculate button.
        
    def convert(self):
        kilo = float(self.kilo_Entry.get())
        
        # Convert kilometers to miles
        miles = kilo * 0.6214
        
        # Display the results in an info dialogue box
        tkinter.messagebox.showinfo('Results',
                                    str(kilo) +
                                    ' kilometers is equal to ' +
                                    str(miles) + ' miles. ')
        
# Creates an instance of the KiloConverterGUI class
kilo_conv = KiloConverterGUI()
````

### Using Labels as Output Fields 13.7
**Concept**: When a StringVar object is associated with a Label widget, the Label widget displays any data that is stored in the StringVar object.

If you don't want to display a separate dialogue box for your program's output, you can use Label widgets in the program's main window to dynamically display output.


````python
# This program converts distance in kilometers
# to miles. The result is displayed in a Label
# on the main window.

import tkinter
import tkinter.messagebox

class KiloConverterGUI:
    def __init__(self):
        # Create the main window widget
        self.main_window = tkinter.TK()
        
        # Create three frames to group widgets
        self.top_frame = tkinter.Frame()
        self.middle_frame = tkinter.Frame()
        self.bottom_frame = tkinter.Frame()
        
        # Create the widgets for the top frame
        self.prompt_label = tkinter.Label(self.top_frame, 
                                          text= 'Enter a distance in kilometers')
        self.kilo_Entry = tkinter.Entry(self.top_frame, 
                                        width= 10)
        
        # Pack the top frame's widgets
        self.prompt_label.pack(side= 'left')
        self.kilo_Entry.pack(side= 'left')
        
        # Create widgets for the middle Frame
        self.desc_label = tkinter.Label(self.middle_frame,
                                        text= 'Converted to miles: ')
        
        # We need a StrVar object to associate with
        # an output Label. Use the objects det method
        # to store a string of blank characters.
        self.value = tkinter.StringVar
        
        # Convert a label and associate it with the
        # StringVar object. Any value stored in the 
        # StringVar object will automatically be displayed
        # In the Label.
        self.miles_label = tkinter.Label(self.middle_frame, 
                                         textvariable= self.value)
        
        # Pack the middle frame's widgets
        self.desc_label.pack(side= 'left')
        self.miles_label.pack(side= 'left')
        
        # Create the button widgets for the bottom frame
        self.calc_button = tkinter.Button(self.bottom_frame,
                                          text= 'Convert',
                                          command= self.convert)
        
        self.quit_button =  tkinter.Button(self.bottom_frame,
                                          text= 'Quit',
                                          command= self.main_window.destroy)
         # Pack the buttons
        self.calc_button.pack(side= 'left')
        self.quit_button.pack(side= 'left')
        
        # Pack the Frames
        self.top_frame.pack()
        self.middle_frame.pack()
        self.bottom_frame.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
    # The convert method is a callback function for
    # the calculate button.
        
    def convert(self):
        kilo = float(self.kilo_Entry.get())
        
        # Convert kilometers to miles
        miles = kilo * 0.6214
        
        # Convert miles to a string and store it 
        # in a StringVar object. This will automatically
        # update the miles_label widget.
        self.value.set(miles)
        
# Creates an instance of the KiloConverterGUI class
kilo_conv = KiloConverterGUI()
````

### Radio Buttons and Check Buttons 13.8
**Concept**: Radio buttons normally appear in groups of two or more, and allow the user to select one of several possible actions. Check button, which may appear alone or in groups, allow the user to make yes/no or on/off selections.

### Radio Buttons
Radio Buttons are useful when you want the user to select one choice from several possible actions. A Radio Button can be selected or deselected. Each Radio Button has a small circle that appears filled when a button is selected, and appears empty when the radio button is deselected.

Only one of the radio button widgets in a container(A Frame), may be selected at any time. Clicking a Radio Button selects it and automatically deselects any other Radio Button, because of this they are said to be mutually exclusive.

The tkinter module provides a class name IntVar that can be used along with the RadioButton widgets. When you create a group of RadioButtons, you associate them with the same IntVar object. You also assign a unique integer value to each RadioButton. When a RadioButton is selected, it stores its unique integer value in the IntVar object.


````python
# This program demonstrates group of RadioButton widgets
import tkinter
import tkinter.messagebox

class KiloConverterGUI:
    def __init__(self):
        # Create the main window widget
        self.main_window = tkinter.TK()
        
        # Create two frames to group widgets
        self.top_frame = tkinter.Frame(self.main_window)
        self.bottom_frame = tkinter.Frame(self.main_window)
        
        # Create an IntVAr object to use with 
        # the RadioButton
        self.radio_var = tkinter.IntVar()
        
        # Set the IntVAr object to 1
        self.radio_var.set(1)
        
        # Create the RadioButton widgets in the top_frame
        self.rb1 = tkinter.Radiobutton(self.top_frame,
                                       text= 'Option 1',
                                       variable= self.radio_var,
                                       value= 1)
        
        self.rb2 = tkinter.Radiobutton(self.top_frame,
                                       text= 'Option 2',
                                       variable= self.radio_var,
                                       value= 2)
        
        self.rb3 = tkinter.Radiobutton(self.top_frame,
                                       text= 'Option 3',
                                       variable= self.radio_var,
                                       value= 3)
        
        # Pack the RadioButtons
        self.rb1.pack()
        self.rb2.pack()
        self.rb3.pack()
        
        self.ok_button = tkinter.Button(self.bottom_frame, 
                                        text= 'OK',
                                        command= self.show_choice)
        
        self.quit_button =  tkinter.Button(self.bottom_frame,
                                          text= 'Quit',
                                          command= self.main_window.destroy)
        
        # Pack the buttons
        self.ok_button.pack(side= 'left')
        self.quit_button.pack(side= 'left')
        
        # Pack the Frames
        self.top_frame.pack()
        self.bottom_frame.pack()
        
        # Enter the tkinter mainloop
        tkinter.mainloop()
        
    # The show_choice method is the callback function for the
    # OK button.
    def show_choice(self):
        tkinter.messagebox.showinfo('Selection', 'You selected option ' +
                                    str(self.radio_var.get()))
        
# Creates an instance of the KiloConverterGUI class
kilo_conv = KiloConverterGUI()
````

A Check buttons appears as a small box with a label next to it.
