### The Program Development Cycle

The process of creating a program that works correctly typically requires the\
five phases shown. The process is known as the program development cycle.

## Program Development Cycle

- **Design the program**. Tons of professional programmers will tell you that a program\
should be designed carefully before the code is written. When writing a program you\
should start by creating a design of the program, and there are plenty of ways to do so.

- **Write the Code**. After designing the program, then you begin writing code in a\
high-level language. While remembering to follow the syntax rules.

- **Correct Syntax Errors**. If the program contains a syntax error, the compiler\
or interpreter will display an error message indicting what the error is. Once the\
error are corrected the code is able to be executed.

- **Test the Program**. Once the code is in an executable form, it is then tested\
to determine whether any logic errors exist. A logic error is a mistake that doesn't\
stop the program form running, but causes it to produce result.

- **Correct Logic Errors**. If the program produces the incorrect results, the\
programmer debugs the code. Debugs means that the programmer finds and corrects logic\
errors in the program. Sometimes during the process the programmer discovers that the programs'\
original design must be changed. in this event, the program development cycle starts over and\
continues until no errors can be found.

### Pseudocode

Because small mistakes in code can cause syntax errors, programmers find it helpful\
write programs in pseudocode before they write the actual code. Pseudocode is fake code,\
it is an informal language that has no syntax rules and is not meant to be compiled or executed.\
Instead, programmers use pseudocode to create models, or "mock-ups," of programs. 

Because programmers don't have to worry about syntax errors while writing pseudocode,\
they can focus all their attention on the programs design.

### Flowcharts

Flowcharting is another tool programmers use to design programs.\
A Flowchart is a diagram that graphically depicts the steps that\
take place in a program. There are three symbols in a flowchart: Ovals,\
Parallelograms, and a rectangle. Each of these represent a step in the program,\
as described:

- The ovals, which appear at the top or bottom of the chart, are called terminal\
symbols. The Start terminal symbol marks the programs starting point, and the\
End terminal symbol marks the programs ending point.

- The parallelograms are uses as input symbols and output symbols. They represent\
steps in which the program reads inputs or displays output.

- Rectangles are used  as processing symbols. The represent steps in which the program\
performs some process on data.

### floating-point and Integer Division

The / operator performs floating division, and the // operator performs integer division.\
Both operators divide one number by another. The difference between them is that the / operator\
gives the result as a floating-point value, and the // operator gives the result as a whole number.\
Let's use interactive mode to demonstrate this:

```

>>>
```

```

```

As you can see the result is 2. The // operator works like this:

- When the result is positive, it is truncated, which means that it's fractional\
part is thrown away.

- When the result is negative, it is rounded away from zero to the nearest integer.

### Operator Precedence

You can write statements that use complex mathematical expressions involving several\
operators. The following statement assigns the sum of 17, the variable x, 21, and the\
variable y to the variable answer:

```python
answer = 17 + x + 21 + y
```

Some expressions are straightforward, however. Consider the following statement:

```python
outcome = 12.0 + 6.0 / 3.0
```

Fortunately, the answer can be predicted because Python follow the same order of operations\
that you learned in math class.

First, operations that are enclosed in parentheses are performed first. Then, when two operators\
share an operand, the operator with the higher precedence is applied first. The precedence of\
the math operators, highest to lowest, are:

1. Exponentation: **
2. Multiplication, division, and remainder: * / // %
3. Addition and subtraction: + -

**Note**: There is an exception to the left-to-right rule. When two ** operators share\
an operand, the operators execute right-to-left. For example,  the expression 2**3**4\
is evaluated as 2**(3**4).

### Grouping with Parentheses

### Mixed-Type Expressions and Data Type Conversion

- When an operation is performed on an int and a float, the int value is temporarily\

### Escape Characters

An escape is a special character that is preceded with a backslash(\), appearing inside\
a string literal. for example, \n is the newline escape character. When the \n escape\
character is printed, it isn't displayed on the screen. Instead, it causes output to\
advance to the next line. For example, look at the following:

```
print('One\nTwo\nThree')
```

When this statement executes, it displays:

```
One
Two
Three
```

| Escape Character |                            Effect                            |
|:----------------:|:------------------------------------------------------------:|
|        \n        |        Causes output to be advanced to the next line.        |
|        \t        | Causes output to skip over the next horizontal tab position. |
|        \'        |          Causes a single quote mark to be printed.           |
|        \"        |          Causes a double quote mark to be printed.           |
|        \\        |         Causes a backslash character to be printed.          |

The \t escape character advances the output to next horizontal position. The following
statements are illustrative:
```
print('Mon\tTues\tWed')
print('Thur\tFri\tSat')
```

The output will look like this:
```
Mode    Tues    Wed
Thurs   Fri     Sat
```

### Formatting Numbers

you might not always be happy with the way that numbers, especially floating-point numbers,\
are displayed on the screen. when a floating-point number is displayed by the print function,\
it can appear with up to 12 significant digits. Because the program displays a dollar amount,\
it would be nice to see the amount rounded to two decimal places. Fortunately, Python gives us\
a way to do just that, and more, with the built-in format function. When you call the built-in\
format function, you pass two arguments to the function: a numeric value and a format specifier.\
The format specifier is a string that contains special characters specifying how the numeric\
value should be formatted. Let's look at an example here:

```
>>> format(12345.6789, '.2f')
12345.68
>>>
```

- The .2 specifies the precision. It indicates that we want to round the number\
to two decimal places.

- The f specifies that the data type of the number we are formatting is a\
floating-point number.

### Formatting in Scientific Notation

if you prefer to display floating-point numbers in scientific notation, you can use the\
letter e or E instead of f.

### Inserting Comma Separators

If you want the number to be formatted with the comma separators, you can insert a comma\
into the format specifier, as show here:

```
>>> format(12345.6789, ',.2f')
12,345.68
>>>
```

Here's an example that specifies the comma separator, but does not specify precision:

```
>>> format(12345.6789, ',f')
12,345.6789000
>>>
```

### Specifying a Minimum Field Width

The format specifier can also include a minimum field width, which is the minimum number of\
spaces that should be used to display the value. The follow example prints the number in a\
field that 12 spaces wide:

```
(With comma)
>>> print('The number is',format(12345.6789, '12,.2f'))
The number is 12,345.68
>>>
```
---------------------------------------------------------
```
(Without comma)
>>> print('The number is',format(12345.6789, '12.2f'))
The number is        12345.68
>>>
```
### Formatting a Floating-Point Number as a Percentage

Instead of using f as the type designator, you can use the % symbol to format a\
floating-point number as a percentage. The % symbol causes the number to be multiplied\
and displayed with a % sign following it.

### Formatting integers

You can use the format function to format integers. There are two differences to keep\
in mind when writing a format specifier  that will be used to format integers:

- You use d as the type designator
- You cannot specify precision

### Named Constants

A named constant is a name that represents a value that cannot be changed during\
the program's execution. Imagine for a moment that you are a programmer working\
at a bank. You are updating an existing program that calculates data pertaining\
to loans, and you see the following:

```
amount =  balnce * 0.069
```

Because someone else wrote the program, you aren't sure what the number 0.069 is. It\
appears to be an interest number, but it could be a number that is used to calculate\
some sort of fee. You simply can't determine the purpose of the number 0.069 by reading\
this line of code. This is an example of magica number. A magic number is an unexplained\
value that appears in a program's code.

Magic numbers can be problematic and for a number of reason. These problems can be addressed\
by using named constants to represent magic numbers. A named constant is a name that represents\
a value that does not change during the programs' execution. The following is how we will
declare named constants in code:

```
INTERSEST_RATE = 0.069
```

4. Total Purchase

A customer in a store is purchasing five items.\
Write a program that ask for the price of each item,\
then displays the subtotal of the sale, the amount of sales, and the total.\
Assume the sales tax is 7 percent.

````python
first_item = float(input("enter the price for 1st item"))
second_item = float(input("enter the price for 2nd item"))
third_item = float(input("enter the price for 3rd item"))
fourth_item = float(input("enter the price for 4th item"))
fifth_item = float(input("enter the price for 5th item"))
sub_total = first_item + second_item + third_item + fourth_item + fifth_item
sales_tax_rate = 7 / 100
sales_tax_amount = sub_total * sales_tax_rate
total = sales_tax_amount + sub_total
print("subtotal:", format(sub_total, ".2f"))
print("sales tax amount:", format(sales_tax_amount, ".2f"))
print("total:", format(total, ".2f"))
````
