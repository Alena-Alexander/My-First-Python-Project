### The Program Development Cycle
The process of creating a program that works correctly typically requires the five phases shown.\
The process is known as the program development cycle.

## Program Development Cycle
- **Design the program**. Tons of professional programmers will tell you that a program should be designed\
carefully before the code is written. When writing a program you should start by creating a design of the program,\
and there are plenty of ways to do so.

- **Write the Code**. After designing the program, then you begin writing code in a high-level language.\
While remembering to follow the syntax rules.

- **Correct Syntax Errors**. If the program contains a syntax error, the compiler or interpreter will display\
an error message indicting what the error is. Once the error are corrected the code is able to be executed.

- **Test the Program**. Once the code is in an executable form, it is then tested to determine whether any logic\
errors exist. A logic error is a mistake that doesn't stop the program form running, but causes it to produce result.

- **Correct Logic Errors**. If the program produces the incorrect results, the programmer debugs the code.\
Debugs means that the programmer finds and corrects logic errors in the program. Sometimes during the process the programmer\
- discovers that the programs' original design must be changed. in this event, the program development cycle starts over and\
- continues until no errors can be found.

### Pseudocode
Because small mistakes in code can cause syntax errors, programmers find it helpful write programs in\
pseudocode before they write the actual code. Pseudocode is fake code, it is an informal language that has no syntax\
rules and is not meant to be compiled or executed. Instead, programmers use pseudocode to create models, or "mock-ups," of programs.\
Because programmers don't have to worry about syntax errors while writing pseudocode,\
they can focus all their attention on the programs design.

### Flowcharts
Flowcharting is another tool programmers use to design programs.\
A Flowchart is a diagram that graphically depicts the steps that take place in a program.\
There are three symbols in a flowchart: Ovals, Parallelograms, and a rectangle.\
Each of these represent a step in the program, as described:
- The ovals, which appear at the top or bottom of the chart, are called terminal symbols.
The Start terminal symbol marks the programs starting point, and the End terminal symbol marks the programs ending point.
- The parallelograms are uses as input symbols and output symbols. They represent steps in which the program reads inputs or displays output.
- Rectangles are used  as processing symbols. The represent steps in which the program performs some process on data.

### Input, Processing, and Output
Input is data that the program receives. When a program receives data, it usually processes it by performing some operation on it.\
The result of the operation is sent out of the program as output

Computer programs typically perform the following three-step process:
1. Input is received.
2. Some process is performed on the input.
3. Output is produced.

Input is any data that the program receives while it is running. one common form of input is data that is typed on the keyboard.\
Once input is received, some process, such as a mathematics calculation is done to it.\
The result of the process is sent out as an output.

### Strings and String Literals
Programs almost always works with data of some type. For example, program 2-1 uses the following three pieces of data:
```
'Kate Austen'
'123 Full Circle Drive'
'Asheville, NC 28899'
```

These three pieces of data are sequences of characters. In programming terms, a sequence of characters that is used as\
data is called a string. When a string appears in the actual code of a program, it is called a string literal.\
String literals must be contained in quote marks. You can enclose string literals in a set of single-quote marks('') or a set of double-quote marks("").

Python also allows you to enclose string literals in triple-quotes (either """ or ''').\
Triple-quoted strings can contain both single and double quotes as part of the string. The following statement shows an example: 
```
print("""I'm reading "Hamlet" tonight.""")
```

This statement will print:
```
I'm reading "Hamlet" tonight.
```

Triple can also be used to surround multiline strings, somthing for which single and double quotes cannot do.\
Here's an example:
```
print("""One
Two
Three""")
```

The statement will print:
```
One 
Two
Three
```

### Comments 2.4

### Variables 2.5
A variable is a name that represents a value stored in the computer's memory.

Programs usually store data in the computer's memory and perform operation's on that data.\
For example, consider the typical online shopping experience: you browse a website and add\
items you want to purchase to the shopping cart. As you add items to the shopping cart, data about\
the items is stored in memory. Then, when you click the checkout button, a program running on the\
website's computer calculates the cost of all the items you have in your shopping cart, the applicable sales tax,\
shipping costs, and the total of all these charges. When the program performs these calculations,\
it stores the results in the computer's memory.

Programs use variables to access and manipulate data that is stored in memory.\
A variable is a name that represents a value in the computers' memory.\
When a variable represents a value in the computers' memory, we say that the variable references the value.

### Creating Variables with Assignment Statements
You use an assignment statement to create a variable and make it reference a piece of data.\
Here is an example of an assignment statement:
```
age = 25
```
After this statement executes, a variable named age will be created, and it will reference the value 25. 

An assignment statement is written in the following format:
variable = expression

the equal sign(=) is known as the assignment operator. In the general format, variable is the name\
of the variable and the expression is the value, or any piece of code that results in a value.\
To experiment with variables, you can type assignment statements in interactive mode, as shown here:
```
>>> width = 10
>>> length = 5
>>>
```
The first statement creates a variable width and assigns it to the value 10 and does the same with length\
and assigns it to 5.\
Nest you use the print function to display the values:
```
>>> print(width)
10
>>> print(length)
5
>>>
```

When you pass a variable as an argument to the print function, you do not enclose the variable name in quote marks.\
To demonstrate why look at the follow interactive mode:
```
>>> print('width')
width
>>> print(width)
10
>>>
```

In first statement, you passed 'width' as an argument to the print function, and the function printed the string width.\
In the second, you pass width (With no quote marks) as an argument to the print function, and the function displayed the value.\

In an assignment statement the that is receiving the assignment must appear on the left side of the = operator,\
and it must not be a variable else you will receive an error.
