### Flowcharts
Flowcharting is another tool programmers use to design programs.\
A Flowchart is a diagram that graphically depicts the steps that take place in a program.\
There are three symbols in a flowchart: Ovals, Parallelograms, and a rectangle.\
Each of these represent a step in the program, as described:
- The ovals, which appear at the top or bottom of the chart, are called terminal symbols.
The Start terminal symbol marks the programs starting point, and the End terminal symbol marks the programs ending point.
- The parallelograms are uses as input symbols and output symbols. They represent steps in which the program reads inputs or displays output.
- Rectangles are used  as processing symbols. The represent steps in which the program performs some process on data.

### floating-point and Integer Division
The / operator performs floating division, and the // operator performs integer division. Both operators divide one number by another. The difference between them is that the / operator gives the result as a floating-point value, and the // operator gives the result as a whole number.
Let's use interactive mode to demonstrate this:
```
>>> 5 / 2
2.5
>>>
```

```
>>> 5 // 2
2
>>>
```
As you can see the result is 2. The // operator works like this:
- When the result is positive, it is truncated, which means that it's fractional part is thrown away.
- When the result is negative, it is rounded away from zero to the nearest integer.

### Operator Precedence
You can write statements that use complex mathematical expressions involving several operators. The following statement assigns the sum of 17, the variable x, 21, and the variable y to the variable answer:
```python
answer = 17 + x + 21 + y
```

Some expressions are straightforward, however. Consider the following statement:
```python
outcome = 12.0 + 6.0 / 3.0
```

Fortunately, the answer can be predicted because Python follow the same order of operations that you learned in math class.
First, operations that are enclosed in parentheses are performed first. Then, when two operators share an operand, the operator with the higher precedence is applied first. The precedence of the math operators, highest to lowest, are:
1. Exponentation: **
2. Multiplication, division, and remainder: * / // %
3. Addition and subtraction: + -

**Note**: There is an exception to the left-to-right rule. When two ** operators share an operand, the operators execute right-to-left. For example,  the expression 2**3**4 is evaluated as 2**(3**4).

```
print('One\nTwo\nThree')
```

When this statement executes, it displays:
```
One
Two
Three
```

### Formatting Numbers
you might not always be happy with the way that numbers, especially floating-point numbers, are displayed on the screen. when a floating-point number is displayed by the print function, it can appear with up to 12 significant digits. Because the program displays a dollar amount, it would be nice to see the amount rounded to two decimal places. Fortunately, Python gives us a way to do just that, and more, with the built-in format function. When you call the built-in format function, you pass two arguments to the function: a numeric value and a format specifier. The format specifier is a string that contains special characters specifying how the numeric value should be formatted. Let's look at an example here:
```
>>> format(12345.6789, '.2f')
12345.68
>>>
```
- The .2 specifies the precision. It indicates that we want to round the number to two decimal places.
- The f specifies that the data type of the number we are formatting is a floating-point number.

### Formatting in Scientific Notation
if you prefer to display floating-point numbers in scientific notation, you can use the letter e or E instead of f.

### Inserting Comma Separators
If you want the number to be formatted with the comma separators, you can insert a comma into the format specifier, as show here:
```
>>> format(12345.6789, ',.2f')
12,345.68
>>>
```

Here's an example that specifies the comma separator, but does not specify precision:
```
>>> format(12345.6789, ',f')
12,345.6789000
>>>
```

### Specifying a Minimum Field Width
The format specifier can also include a minimum field width, which is the minimum number of spaces that should be used to display the value. The follow example prints the number in a field that 12 spaces wide:
```
(With comma)
>>> print('The number is',format(12345.6789, '12,.2f'))
The number is 12,345.68
>>>
```
---------------------------------------------------------

Instead of using f as the type disignataor, you can use the % symbol to format a floating-point number as a percentage. The % symbol causes the number to be multiplied and displayed with a % sign following it.

### Named Constants
A named constant is a name that represents a value that cannot be changed during the program's execution.
Imagine for a moment that you are a programmer working at a bank. You are  updating an existing program that calculates data pertaining to loans, and you see the following:
```
amount =  balnce * 0.069
```

Because someone else wrote the program, you aren't sure what the number 0.069 is. It appears to be an interest number, but it could be a number that is used to calculate some sort of fee. You simply can't determine the purpose of the number 0.069 by reading this line of code. This is an example of magica number. A magic number is an unexplained value that appears in a program's code.

Magic numbers can be problematic and for a number of reason. These problems can be addressed by using named constants to represent magic numbers. A named constant is a name that represents a value that does not change during the programs' execution. The following is how we will declare named constants in code:
```
INTERSEST_RATE = 0.069
```
