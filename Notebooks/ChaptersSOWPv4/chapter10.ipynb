# Classes and Object-Oriented Programming
### Procedural and Object-Oriented Programming 10.1

**Concept**: Procedural Programming is a method of writing software.\
It is a programming practice centered on the procedures or actions\
that take places in a program. Object-Oriented Programming is centered on objects.\
Object are created from abstract data types that encapsulates data and functions together.

There are primarily two methods of programming in use today: procedural and object-oriented.\
The earliest programming languages were procedural, meaning a program made of one or more procedures.\
You can think of a procedure simply as a function that performs a specific task such as gathering input\
from the use, performing calculations, reading or writing files, displaying output, and so on.\
The programs that you've written have been procedural. 

Procedures operate on data items that are separate from procedures. The focus of procedural programming\
is on the creation of procedures that operate on the programs' data.

Procedural programming is centered on procedures(functions), Object-Oriented Programming(OOP)\
is centered on objects. An object is a software that contains both data and procedures.\
The data contained in an object is known as its data attributes. An objects data attributes are\
simply variables that reference data. The procedures that an object performs are known as methods.\
An objects methods are function that perform operations on data attributes.\
The object is a self-contained unit that consist of data attributes and methods that operate on them.

OOP addresses the problem of code and data separation through encapsulation and data hiding.\
Encapsulation refers to the combining of data and code into a single object. Data hiding refers\
to the objects ability to hide its data from code that is outside the object.\
Only the objects methods may directly access and make changes to the objects data.

### Object Reusability

The use of OOP has also been encouraged by the trend of Object Reusability.\
An object is not a stand-alone program, but is used by programs that need its services.

For example, Sarah is a programmer who had developed a set of objects for rendering 3D images.\
Tom, is writing a program for an architectural firm, can use Sharon's objects to perform the 3D rendering.

### Classes 10.2

**Concept**: A class is code that specifies the data attributes and methods\
for a particular type of object.

Before an object can be created, it must be designed by a program to determine\
what data attributes and method are necessary, then a class is created.\
You can think of a class as a blueprint.

 A class is description of an objects characteristics. Each object that is created from a class\
 is called an instance of the class.


### Class Definitions

To create a class, you write a class definition. A class definition is a set of statements\
that defines a class's methods and data attributes. 

Here's an example of a class:

````python
import random

class Coin:
    """
    Description
    ==========
    Simulates a coin being flipped.
    
    Usage
    =====
    >>> my_coin = Coin()
    """
    
    def __init__(self):
        """
        Initializes the sideup data attribute with 'Heads'.
        """
        self.sideup = 'Heads'
        
    
    
    def toss(self):
        """
        Generate a random number if the number is 0,
        the attribute is set to 'Heads', if the number is 1 it's set to 'Tails'.
        
        :return: str
        """
        
        # The Toss method generates a random number
        # in the range of 0 through 1. If the number 
        # is 0, then sideup is set to 'Heads'. 
        # Otherwise, sideup is set to 'Tails'.
        if random.randint(0, 1) == 0:
            self.sideup = 'Heads'
        else:
            self.sideup = 'Tails'
            
    
    def get_sideup(self):
        """
        Return the value referenced by the sideup attribute.
        
        :return: str
        """
        return self.sideup
````

The same rule that apply to variable names also apply to class names.\
However, notice how we started the class name, Coin, with an uppercase letter.\
This is not required, but it is a widely used convention among programmers.\
The Coin class has three methods:

- The __init__ method 
- The Toss method
- The get_sideup method\
Notice that these methods look like any other function definition in Python.

And notice that each method has a parameter variable named self:
```
Line 11: def __init__(self):
Line 19: def toss(self):
Line 28: def get_sideup(self):
```

The self parameter is required in every method of a class. When a method executes,\
it must have a way of knowing which object's data attributes it is supposed to operate on.\
That's when the self parameter comes in. When a method is called, Python makes the self\
parameter reference the specific object the method is supposed to operate on. 

Most Python classes have a method named __init__ which is automatically created when an instance\
of a class is created in memory. The __init__ method is commonly known as the initializer method\
because it initializes the objects data attributes. Immediately, after an object is created in memory,\
the __init__ executes, and the self parameter is automatically assigned the object that was created.

As result each object that we create from the Coin class will initially have a side up attribute\
that is set too heads.

````python
import random

# The coin class simulates a coin that can 
# be flipped

class Coin:
    """
    Description
    ==========
    Simulates a coin being flipped.
    
    Usage
    =====
    >>> my_coin = Coin()
    """
    
    def __init__(self):
        """
        Initializes the sideup data attribute with 'Heads'.
        """
        self.sideup = 'Heads'
        
    
    
    def toss(self):
        """
        Generate a random number if the number is 0,
        the attribute is set to 'Heads', if the number is 1 it's set to 'Tails'.
        
        :return: str
        """
        
        # The Toss method generates a random number
        # in the range of 0 through 1. If the number 
        # is 0, then sideup is set to 'Heads'. 
        # Otherwise, sideup is set to 'Tails'.
        if random.randint(0, 1) == 0:
            self.sideup = 'Heads'
        else:
            self.sideup = 'Tails'
            
    
    def get_sideup(self):
        """
        Return the value referenced by the sideup attribute.
        
        :return: str
        
        """
        return self.sideup
    
# The main function
def main():
    # Create an object from the coin class
    my_coin = Coin()
    
    # Display the side of your coin that is facing up.
    print('This side is up: ', my_coin.get_sideup())
    
    # Toss the coin
    print('I am tossing the coin...')
    my_coin.toss()
    
    # Display the side of the coin that is facing up.
    print('This side is up: ', my_coin.get_sideup())
    
# Call the main function
main()
````

The expression Coin() that appears on the right side of the = operator causes two things to happen:

1. An object is created from memory in the coin class.
2. The Coin class __init__ is executed, and self parameter is automatically\
set to the object that was created. As a result the object sideup attribute\
is assigned to string 'Heads'.

### The BankAccount Class
Objects created from this class will simulate bank accounts, allowing us to
have a starting balance, make deposits, make withdrawals, and get the current balance.

````python
class BankAccount:
    """
    Description
    ============
    Simulates a bank account.
    
    Usage
    =====
    >>> savings = bankaccount.BankAccount(start_bal)
    >>> import bankaccount
    """
    
    def __init__(self, bal):
         """
         Initializes the __balance attribute with bal(balance).
         
         :param bal: Starting balance that deposits and withdrawals will be made to.
         """
         self.__balance = bal
         
    def deposit(self, amount):
        """
        Adds an amount to the __balance attribute.
        
        :param self: The object that represents __balance.
        :param amount: The currency that will be added to __balance
        :return: none
        """
        self.__balance += amount
          
    def withdrawal(self, amount):
        """
        Withdraws amount from __balance attribute, if condition is met.
        
        :param self: The object that represents __balance.
        :param amount: The currency that will be removed from the __balance
        :return: none
        """
        if self.__balance >= amount:
            self.__balance -= amount
        else:
            print('ERROR: Insufficient Funds')
    
    def get_balance(self):
        """
        Returns the account balance.
        
        :param self: The object that represents __balance.
        :return: The current balance.
        """
        return self.__balance
````

````python
# This program demonstrates the BankAccount class
import bankaccount

def main():
    # Get the starting balance
    start_bal = input('Enter your starting balance: ')
    
    # Get a bankaccount object
    savings = bankaccount.BankAccount(start_bal)
    
    # Deposit the users paycheck
    pay = input('How much were you paid this week: ')
    print('I will deposit that into your account.')
    savings.deposit(pay)
    
    # Display the balance
    print('Your account balance is $', 
          format(savings.get_balance(), ',.2f'), 
          sep='')
    
    # Get the amount of withdraw
    cash = input('How much would you like to withdraw? ')
    print('I will withdraw that from your account.')
    savings.withdrawal(cash)
    
    # Display the balance
    print('Your account balance is $', 
          format(savings.get_balance(), ',.2f'), 
          sep='')
    
# Call the main function
main()
````

### The __str__ Method
Often we need to display a message that indicates an objects state. An objects state\
is simply the values of the objects attributes at any given moment.

Displaying an objects state is a common task. It is so common that many programmers equip\
their classes with a method that returns a string containing the objects state.

````python
class BankAccount:
    """
    Description
    ============
    Simulates a bank account.
    
    Usage
    =====
    >>> savings = bankaccount.BankAccount(start_bal)
    >>> import bankaccount
    """
    
    def __init__(self, bal):
         """
         Initializes the __balance attribute with bal(balance).
         
         :param bal: Starting balance that deposits and withdrawals will be made to.
         """
         self.__balance = bal
         
    def deposit(self, amount):
        """
        Adds an amount to the __balance attribute.
        
        :param self: The object that represents __balance.
        :param amount: The currency that will be added to __balance
        :return: none
        """
        self.__balance += amount
          
    def withdrawal(self, amount):
        """
        Withdraws amount from __balance attribute, if condition is met.
        
        :param self: The object that represents __balance.
        :param amount: The currency that will be removed from the __balance
        :return: none
        """
        if self.__balance >= amount:
            self.__balance -= amount
        else:
            print('ERROR: Insufficient Funds')
    
    def get_balance(self):
        """
        Returns the account balance.
        
        :param self: The object that represents __balance.
        :return: The current balance.
        """
        return self.__balance
    
    def __str__(self):
        """
        Returns the __balance attribute indicating its state.
        
        :return: float
        """
        return 'The balance is $', format(self.__balance, ',.2f')
````

````python
class Cellphone:
    """
    Description
    ===========
    Holds data about the cellphone.
    
    Usage
    =====
    >>> phone = cellphone.Cellphone(man, mod, retail)
    """

    def __init__(self, manufact, model, price):
        """
        Initializes the three attributes(__manufact, __model, and __price) with manufact, model, price.
        
        :param manufact: The manufact of the cellphone.
        :param model: The model of the cellphone.
        :param price: The cost of the cellphone.
        """
        self.__manufact = manufact
        self.__model = model
        self.__retail_price = price
        
    def set_manufact(self, manufact):
        """
        Accepts an argument for the phone manufacturer.
        
        :param manufact: The manufact of the cellphone.
        :return: none
        """
        self.__manufact = manufact
    
    def set_model(self, model):
        """
        Accepts an argument for the phone's model number.
        
        :param model: The model of the cellphone.
        :return: none
        """
        self.__model = model
    
    def set_retail_price(self, price):
        """
        Accepts an argument for the phone's retail price.
        
        :param price: The cost of the cellphone.
        :return: none
        """
        self.__retail_price = price
        
    def get_manufact(self):
        """
        Returns the phone's manufacturer.
        
        :return: str
        """
        return self.__manufact
    
    def get_model(self):
        """
        Returns the phone's model number.
        
        :return: int
        """
        return self.__model
    
    def get_retail_price(self):
        """
        Returns the phone's retail price.
        
        :return: int
        """
        return self.__retail_price
````

## Accessor and Mutator Methods
A method returns a value from the class attribute but does not change\
it is known as an accessor method. Accessor methods provide a safe way for\
code outside the class to retrieve the values of attributes, without exposing\
the attributes in a way that they could be changed.\
The get_manufact, get_model, and get_retail_price method are accessor methods.

A method that stores a value in a data attribute or changes the value of a data attribute\
is known as a mutator method. Mutator methods can control the way a classes data attributes\
are modified. When code outside the class needs to change the value of an object's data attribute,\
it typically calls a mutator, and passes the new value as an argument.\
The set_manufact, set_model, and set_retail_price methods are mutator methods.

### Passing Objects as Arguments
When you are developing applications that work with objects, you often need to write functions\

When you pass an object as an argument, the thing that is passed into the parameter variable\

an argument has access to the actual object. For example, look at the following flip method:

```

    coin_obj.toss()
```

### Finding the Classes in a Program
When developing an object-oriented program, one of your first task is to\
identify the class that you will need to create. Typically, your goal is to\
identify the different types of real-objects that present in the problem,\
then create classes for those types of objects within your application.

Software professionals have developed numerous techniques for finding classes in a given problem.\
One simple and popular technique involves the following steps:

1. Get a written description of the problem domain.
2. Identity all the nouns(including pronouns and noun phrases)in the description. Each of these is a potential class.
3. Refine the list to include only the classes that are relevant to the problem.

### Writing a Description of the Problem Domain
The problem domain is the set of real-world objects, parties, major events related to the problem.\
If you adequately understand the nature of the problem you are trying to solve,\
you can write a description of the problem domain yourself.\
Though if you don't understand the nature of the problem,\
you should have an expert write it for you.

For example, suppose we are writing a program that the manager of Joe's Automotive Shop\
will use to print service quote for customer's. Here's a description that an expert, has written: 

```

Porsche, and BMW. When a customer brings a car to the shop, the manager gets the customer's
name, address, and telephone number. The manager then determines the make, model, and year
of the car and gives the customer a service quote. The service qoute shows the estimated parts
```

The problem domain description should include any of the following:

- Physical objects such as vehicles, machines, or products
- Any role played by a person, such as a manager, employee, or in this case a service quote.
- Record Keeping items, such as customer histories and payroll records

### Identifying a Class's Responsibilities
Once the classes have been identified, the next step is to identify their responsibilities.\
A class's responsibilities are:

- The things that the class is responsible for knowing
- The action that the class is responsible for doing

When you've identified what the class is responsible for you have identified\
the class's attributes. Same as when you identify the actions that a class is\
responsible for, the methods.

### The Customer Class Ex.
In context of the problem domain, what must the customer Class know?
The description mentions the following items, which are all data attributes of a customer:

- the customers name
- the customers address
- the customers telephone number

In context of the problem domain, what must the Customer Class do?
The only obvious actions are:

- initializes an object of the Customer Class
- set and return the customers name
- set and return the customers address
- set and return the customers telephone number

### This is only the Beginning
It's important to realize that designing an object-oriented application is an iterative process.\
It may take you several attempts to identify all the classes that you'll need and determine all\
their responsibilities. As the design proces unfolds, you will gain a deeper understanding of\
the problem, and consequently you will see ways to improve the design.

1. Pet Class

````python
class Pet:
    """
    Description
    ==========
    Holds data about the pet.
    
    Attributes
    ==========
    name: Container for str
    animal_type: Container for str
    age: Container for int
    
    Usage
    =====
    >>> ThePet = Pet(PetName, PetType, PetAge)
    """
    
    def __init__(self, name: str, animal_type: str, age: int):
        """
        Initializes the three attributes(__name, __animal_type, and __age) with name, animal_type, age.
        
        :param name: Contains the name of the pet.
        :param animal_type: Contains the type of pet.
        :param age: Contains the age of the pet.
        """
        self.__name = name
        self.__animal_type = animal_type
        self.__age = age
        
    def set_name(self, name: str):
        """
        Accepts an argument for the name attribute.
        
        :param name: The name of the pet.
        :return: none
        """
        self.__name = name
    
    def set_animal_type(self, animal_type: str):      
        """
        Accepts an argument for the animal type attribute.
        
        :param animal_type: The type of animal.
        :return: none
        """
        self.__animal_type = animal_type
        
    def set_age(self, age: str):
        """
        Accepts an argument for the age attribute.
        
        :param age: The age of the animal.
        :return: none
        """
        self.__age = age
        
    def get_name(self):
        """
        Returns the name attribute. 
        
        :return: str
        """
        return self.__name
    
    def get_animal_type(self):
        """
        Returns the animal type attribute. 
        
        :return: str
        """
        return self.__animal_type
    
    def get_age(self):
        """
        Returns the age attribute.
        
        :return: int
        """
        return self.__age
    
    
def main():
    
    PetName = input('Enter pet name: ')
    PetType = input('Enter the type of pet: ')
    PetAge = int(input('Enter the pets age: '))
    
    ThePet = Pet(PetName, PetType, PetAge)
    
    print('Name: ', ThePet.get_name())
    print('Type: ', ThePet.get_animal_type())
    print('Age: ', ThePet.get_age())
    
main()
````

2. Car Class

````python
class Car:
    """
    Description
    ===========
    Holds the data of the car, and adjust the speed of it.
    
    Attributes
    ==========
    year_model: Container for str
    make: Container for str
    speed: Container for int
    
    Usage
    =====
    >>> TheCar = Car(yearModel, make, speed)
    """
    
    def __init__(self, year_model: str, make: str, speed: int):
        """
        Initializes the three attributes(__year_model, __make, and __speed) with year_model, make, and speed.
        
        :param year_model: Contains the cars year model.
        :param make: Contains the make of the car.
        :param speed: Contains the speed of the car
        """
        self.__year_model = year_model
        self.__make = make
        self.__speed = speed
        
    def accelerate(self):
        """
        Adds +5 to the speed attribute.
        
        :return: none
        """
        self.__speed += 5
        
    def brake(self):
        """
        Subtracts -5 from the speed attribute.
        
        :return: none
        """
        self.__speed -= 5
        
    def get_speed(self):
        """
        Return the speed attribute.
        
        :return: int
        """
        return self.__speed
    
def main():
    
    yearModel = input('Enter the year model of the car: ')
    make = input('Enter the make of the car: ')
    speed = int(input('Enter the speed of the car: '))
    
    TheCar = Car(yearModel, make, speed)
    
    NumOfCalls = 5
    
    for n in range(NumOfCalls):
        TheCar.accelerate()
        print('The current speed of the car: ', TheCar.get_speed())  
        TheCar.brake()
        print('The current speed of the car: ', TheCar.get_speed())
        
main()
````

4. Employee Class

````python
class Employee:
    """
    Description
    ===========
    Holds data about an employee.
    
    Attributes
    ==========
    name: Container for str
    id_number: Container for int
    department: Container for str
    job_title: Container for str
    
    Usage
    =====
    >>> __Employee_ = Employee(name, id_number, department, job_title)
    """
    
    def __init__(self, name: str, id_number: int, department: str, job_title: str):
        """
        Initializes four attributes(__name, __id_number, __department, __job_title) with name, id_number, department, and job_title.
        
        :param name: Contains the employee's name.
        :param id_number: Contains the employee's ID number.
        :param department: Contains the department an employee works in.
        :param job_title: Contains the employee's job title.
        """
        self.__name = name
        self.__id_number = id_number
        self.__department = department
        self.__job_title = job_title
        
    def set_name(self, name: str):
        """
        Accepts an argument for the name attribute.
        
        :param name: Name of an employee.
        :return: None
        """
        self.__name = name
        
    def set_id_num(self, id_number: int):
        """
        Accepts an argument for the ID number attribute.
        
        :param id_number: An employee's ID.
        :return: None
        """
        self.__id_number = id_number
    
    def set_department(self, department: str):
        """
        Accepts an argument for the department attribute.
        
        :param department: An employee's department.
        :return: None
        """
        self.__department = department
        
    def set_job_title(self, job_title: str):
        """
        Accepts an argument for the job title attribute. 
        
        :param job_title: An employee's job title.
        :return: None 
        """
        self.__job_title = job_title
        
    def get_all(self):
        """
        Returns the attributes name, ID number, department, and job title.
        
        :return: str, int, str, str
        """
        return f"Employee({self.__name, self.__id_number, self.__department, self.___job_title})"
    
    
for n in range(0, 3):
    name = input("Enter an employee's name: ")
    id_number = int(input("Enter the employee's ID number: "))
    department = input("Enter the employee's department: ")
    job_title = input("Enter the job title: ")
        
    __Employee_ = Employee(name, id_number, department, job_title)
        
    print(__Employee_.get_all())
        
````
