# List Commands

### Task

Consider a list (list = []). You can perform the following commands:

1. insert `i` `e`: Insert integer `e` at position `i`.
1. print: Print the list.
3. remove e: Delete the first occurrence of integer `e`.
4. append e: Insert integer `e` at the end of the list.
5. sort: Sort the list.
6. pop: Pop the last element from the list.
7. reverse: Reverse the list.

Initialize your list and read in the value of `n` followed by `n` lines of 
commands where each command will be of the `7` types listed above. Iterate 
through each command in order and perform the corresponding operation on your list.

```input.txt
Example 
 N = 4
 append 1
 append 2
 insert 3 1
```

#### Commands In the Input

- Append 1: Append 1 to the list,arr = `[1]` .
- Append 2: Append 2 to the list, arr = `[1,2]` .
- Insert 3 1: Insert 1 at index 3, arr = `[1,3,2]` .
- Print: Print the array. 

#### Output:

```output.txt
[1,3,2]
```

### Input Format

The first line contains an integer, `n`, denoting the number of commands. 
Each line `i` of the `n` subsequent lines contains one of the commands described above.

#### Constraints
- The elements added to the list must be integers.

###  Output Format
For each command of type `print`, print the list on a new line.

### Sample input()

```input.txt
12
insert 0 5
insert 1 10
insert 0 6
print
remove 6
append 9
append 1
sort
print
pop
reverse
print
```


Sample output()

```output.txt
[6, 5, 10]
[1, 5, 9, 10]
[9, 5, 1]
```



``````````python
# TODO: Input Commands
commands = """
12
insert 0 5
insert 1 10
insert 0 6
print
remove 6
append 9
append 1
sort
"""

# create a container for the commands
command_list = []
target_sentinel = 0  # initialized as 0
stop_sentinel = 0  # initialized as 0
command_input = None  # Individual command returned from the `input()` class

# We now need to get the commands from the user
while True:
    # if the first command has not been received, target sentinel should equal 0.
    if target_sentinel == 0:
        # we have received the first command
        target_sentinel = int(input("Enter the number of commands you are sending!"))
        continue

    # if the stop sentinel equals the sentinel value then all commands have been received.
    if stop_sentinel == target_sentinel:
        break

    # Capture the next command
    command_input = input(
        "Paste or Enter you commands one at a time and enter -1 to EXIT."
    )

    # if the command passed is the sentinal stop value `-1`, we break the loop
    if command_input == "-1" or command_input == -1:
        break

    # split the command string by space separated words and symbols,
    # convert the list to a tuple and add the tuple to the command list
    command_list.append(tuple(command_input.split(" ")))
    # increment our stop sentinel
    stop_sentinel += 1

if command_input == "-1" or command_input == -1:
    print("Thank you for playing......")
    exit(0)

print(command_list)

# create a list container
lst = []

# Iterate over the command list collection
for i, command in enumerate(command_list):
    print(i, command)
    # 1. insert e at i: Insert integer ``e`` at position/index ``i``.
    if command[0] == "insert":
        lst.insert(int(command[1]), int(command[2]))

    # 2. print: Print the list.
    if command[0] == "print":
        print(lst)

    # 3. remove e: Delete the `````````first````````` occurrence of integer ``e``.
    if command[0] == "remove":
        try:
            lst.remove(int(command[1]))
        except ValueError as e:
            print(f"{command[1]} not in the list.")
        except NameError as e:
            print(f"{command[1]} is not the correct name.")
        except Exception as e:
            print(f"{command[1]} not in the list.")
    # # 4. append e: Insert integer ``e`` at the end of the list.
    if command[0] == "append":
        lst.append(int(command[1]))

    # # 5. sort: Sort the list.
    if command[0] == "sort":
        lst.sort()

    # # 6. pop: Pop the last element from the list.
    if command[0] == "pop":
        lst.pop()

    # # 7. reverse: Reverse the list.
    if command[0] == "reverse":
        lst.reverse()

print(f"Final List: {lst}")
``````````
