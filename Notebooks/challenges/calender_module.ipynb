# Task
You are given a date. Your task is to find what the day is on that date.

### Input Format
A single line of input containing the space separated month, day and year, respectively, in `MM DD YYYY` format.

### Constraints
```
`2000 < year < 3000`
```

### Output Format
Output the correct day in capital letters.


### Sample Input
```
08 05 2015
```

### Sample Output
```
WEDNESDAY
```

### Explanation
The day on August `5th 2015` was `WEDNESDAY`.

````python
import calendar

# This Dictionary containing the days in a week
days = {
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday",
}

# This iterates over the dict and outputs a day depending on the date being input
month, day, year = 8, 5, 2015  # The month, day, and year
day = calendar.weekday(
    year, month, day
)  # Returns the exact day of the week depending on the date, month, and year given to it
print(day)
print(days[day].upper())
````
