# String Validators
Python has built-in string validation methods for basic data. It can check if a string is composed of alphabetical characters, alphanumeric characters, digits, etc.

`str.isalnum() `
This method checks if all the characters of a string are alphanumeric (a-z, A-Z and 0-9).
```
>>> print 'ab123'.isalnum()
True
>>> print 'ab123#'.isalnum()
False
```

`str.isalpha(`) 
This method checks if all the characters of a string are alphabetical (a-z and A-Z).
```
>>> print 'abcD'.isalpha()
True
>>> print 'abcd1'.isalpha()
False
```

`str.isdigit()`
This method checks if all the characters of a string are digits (0-9).
```
>>> print '1234'.isdigit()
True
>>> print '123edsd'.isdigit()
False
```

`str.islower()` 
This method checks if all the characters of a string are lowercase characters (a-z).
```
>>> print 'abcd123#'.islower()
True
>>> print 'Abcd123#'.islower()
False
```

`str.isupper()` 
This method checks if all the characters of a string are uppercase characters (A-Z).
```
>>> print 'ABCD123#'.isupper()
True
>>> print 'Abcd123#'.isupper()
False
```

# Task
You are given a string `S`. 
Your task is to find out if the string `S` contains: alphanumeric characters, alphabetical characters, digits, lowercase and uppercase characters.

### Input Format
A single line containing a string `S`.

### Constraints
```
0 < len(S) < 1000
```

### Output Format
In the first line, print `Tru`e if `S` has any alphanumeric characters. Otherwise, print `False`. 
In the second line, print `True` if `S` has any alphabetical characters. Otherwise, print `False`. 
In the third line, print `True` if `S` has any digits. Otherwise, print `False`. 
In the fourth line, print `True` if `S` has any lowercase characters. Otherwise, print `False`. 
In the fifth line, print `True` if `S` has any uppercase characters. Otherwise, print `False`.


### Sample Input
```
qA2
```

### Sample Output
```
True
True
True
True
True
```




````python
def alnum(string__: str) -> bool:
    """This function tests to see if the ``string__`` contains alphanumeric characters (a-z, A-Z and 0-9),
    and returns `True` if the string contains alphanumeric characters, or `False` if it does not contain any alphanumeric characters.
    
    Usage:
    =====
    
    >>> print(alnum("1. This is a test string"))
    >>> True
     
    :param string__: (str): string context to be evaluated, if it contains alphanumeric characters (a-z, A-Z and 0-9).
    :return: (bool): returns a boolean value indicating what for true and what for false.
    """
    # For-Loop: given a ``range(n)`` iterator of size ``n``, iterate over the list of integers (``a``) from (``0, ..., n``).
    for a in range(len(string__)):
        # Check to see if the character at index ``a`` is an alphanumeric character (a-z, A-Z and 0-9).
        if string__[a].isalnum():  # The index is incrementing from: (0, 1, 2, ..., n); where ``n`` is the size of the container.
            # Returns `True` if the character is an alphanumeric character (a-z, A-Z and 0-9).
            return True
    # Otherwise, return `False` if the character is not an alphanumeric character (a-z, A-Z and 0-9).
    return False

def alpha(string__: str) -> bool:
    """This function tests to see if the ``string__`` contains alphabetical characters (a-z and A-Z),
    and returns `True` if the string contains alphabetical characters, or `False` if it does not contain any alphabetical characters.
    
    Usage:
    =====
    
    >>> print(alph("1. This is a test string"))
    >>> True
     
    :param string__: (str): string context to be evaluated, if it contains alphabetical characters (a-z and A-Z).
    :return: (bool): returns a boolean value indicating what for true and what for false.
    """
    # For-Loop: given a ``range(n)`` iterator of size ``n``, iterate over the list of integers (``alpha``) from (``0, ..., n``).
    for alph in range(len(string__)):
        # Check to see if the character at index ``alpha`` is an alphabetical character (a-z and A-Z).
        if string__[alph].isalpha(): # The index is incrementing from: 0, 1, 2, ..., n; where ``n`` is the size of the container.
            # Returns `True` if the character is an alphabetical character (a-z and A-Z).
            return True
        # Otherwise, return `False` if the character is not an alphabetical character (a-z and A-Z).
    return False

def digit(string__: str) -> bool:
    """This function tests to see if the ``string__`` contains digits (0-9),
    and returns `True` if the string contains digits, or `False` if it does not contain any digits.
    
    Usage:
    =====
    
    >>> print(digit("1. This is a test string"))
    >>> True
     
    :param string__: (str): string context to be evaluated, if it contains digits (0-9).
    :return: (bool): returns a boolean value indicating what for true and what for false.
    """
    # For-Loop: given a ``range(n)`` iterator of size ``n``, iterate over the list of integers (``d``) from (``0, ..., n``).
    for d in range(len(string__)):
        # Check to see if the character at index ``d`` is a digit (0-9).
        if string__[d].isdigit(): # The index is incrementing from: 0, 1, 2, ..., n; where ``n`` is the size of the container.
            # Returns `True` if the character is a digit (0-9).
            return True
        # Otherwise, return `False` if the character is not a digit (0-9).
    return False
    
def lower(string__: str) -> bool:
    """This function tests to see if the ``string__`` contains lowercase characters (a-z),
    and returns `True` if the string contains lowercase characters, or `False` if it does not contain any lowercase characters.
    
    Usage:
    =====
    
    >>> print(lower("1. This is a test string"))
    >>> True
     
    :param string__: (str): string context to be evaluated, if it contains lowercase characters (a-z).
    :return: (bool): returns a boolean value indicating what for true and what for false.
    """
    # For-Loop: given a ``range(n)`` iterator of size ``n``, iterate over the list of integers (``l``) from (``0, ..., n``).
    for l in range(len(string__)):
        # Check to see if the character at index ``l`` is a lowercase character (a-z).
        if string__[l].islower(): # The index is incrementing from: 0, 1, 2, ..., n; where ``n`` is the size of the container.
            # Returns `True` if the character is a lowercase character (a-z).
            return True
        # Otherwise, return `False` if the character is not a lowercase character (a-z).
    return False

def upper(string__: str) -> bool:
    """This function tests to see if the ``string__`` contains uppercase characters (A-Z),
    and returns `True` if the string contains uppercase characters, or `False` if it does not contain any uppercase characters.
    
    Usage:
    =====
    
    >>> print(upper("1. This is a test string"))
    >>> True
     
    :param string__: (str): string context to be evaluated, if it contains auppercase characters (A-Z).
    :return: (bool): returns a boolean value indicating what for true and what for false.
    """
    # For-Loop: given a ``range(n)`` iterator of size ``n``, iterate over the list of integers (``u``) from (``0, ..., n``).
    for u in range(len(string__)):
        # Check to see if the character at index ``u`` is an uppercase character (A-Z).
        if string__[u].isupper(): # The index is incrementing from: 0, 1, 2, ..., n; where ``n`` is the size of the container/string.
            # Returns `True` if the character is an uppercase character (A-Z)*.
            return True
        # Otherwise, return `False` if the character is not an uppercase character (A-Z).
    return False
    
````

````python
string = "45"
print(alnum(string))
print(alpha(string))
print(digit(string))
print(lower(string))
print(upper(string))
````
